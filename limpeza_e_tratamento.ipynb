Carregando o arquivo CSV para dentro do data frame

```python
import pandas as pd
import numpy as np

df_steam = pd.read_csv('SteamGames.csv')
#Criando uma KPI
df_steam['Total_de_Vendas'] = (df_steam['PositiveReview'] + df_steam['PositiveReview']).astype(float)
df_steam.head()
```

Sustitutindo dados vazios por dados consistentes.

```python
substituicao_dados = {
    'Tags': 'Indie',
    'OsRequirement': 'Windows 11',
    'MemoryRequirement': '8 GB RAM',
    'CpuRequirement': 'Intel Core i5-3570K or AMD Ryzen 3 2200G'
}

df_steam.fillna(value=substituicao_dados, inplace=True)
```

```python
df_steam.info()
```

Preenchendo os dados da coluna Genres vazios.


```python
df_steam['Genres'] = df_steam['Genres'].fillna('Indie')
```

Organizando as datas de lançamento

```python
df_steam['ReleaseDate'] = pd.to_datetime(df_steam['ReleaseDate'], format='%b %d, %Y', errors='coerce')
df_steam.dropna(subset=['ReleaseDate'], inplace=True)
df_steam.info()

```

Filtrando dados de jogos AAA(Grandes Lançamentos)

```python
# Converte a coluna price para conseguir ser lido.
df_steam['price'] = pd.to_numeric(
    df_steam['price'].astype(str)
        .str.replace(r'[^0-9.,]', '', regex=True)   
        .str.replace(',', '.', regex=False),        
    errors='coerce'                                 
).fillna(0.0)
condicao_year = (df_steam['ReleaseDate'].dt.year >= 2020) & (df_steam['ReleaseDate'].dt.year <= 2025)
condicao_price = (df_steam['price'] > 49.99)
condicao_review_player = (df_steam['PositiveReview'] > df_steam['NegativeReview'])

df_gamefilter_aaa = df_steam.loc[condicao_year & condicao_price & condicao_review_player]

print(f"Total de jogos filtrados (AAA): {len(df_gamefilter_aaa)}")
```

Teste de agrupamento

```python
teste_agrupamento = df_gamefilter_aaa.groupby('Publishers').agg({'Name': 'count'})
print(teste_agrupamento)
```

Recolhendo o total de vendas das 50 distribuidoras que mais venderam entre 2020 e 2025

```python
pd.options.display.float_format = '{:,.2f}'.format
recolhimento = {
    'price': 'mean',
    'Total_de_Vendas': 'sum',}
df_gamefilter_aaa.groupby('Publishers').agg(recolhimento).sort_values(by='Total_de_Vendas', ascending=False).head(50)
```

Salvando a otimização feita.

```python
df_steam.to_csv('SteamGames_clean.csv', index=False)
```
